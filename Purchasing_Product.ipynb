##Read Inventory

```python
import json
```

```python
fd = open("record.json",'r')
r = fd.read()
fd.close()

records = json.loads(r)
```

##Purchase Product Which You Want


```python
print("WELCOME")
ui_prod  = str(input("Enter the product_Id: "))
ui_quant = int(input("Enter the quantity: "))

if ui_quant == 0 :
  print('The product is not available')
elif (ui_prod not in records.keys()):
  print("Sorry,You entered wrong ID.")
elif (ui_quant > records[ui_prod]['qn']):
  print("Your Entered Quantity exceeds than Record Quantity")  
else :
  print('the product is available')  
print('********************')
print("Product: ", records[ui_prod]['name'])
print("Price: ", records[ui_prod]['pr'])
print('********************')
print("Billing Amount: ", records[ui_prod]['pr'] * ui_quant)

records[ui_prod]['qn'] = records[ui_prod]['qn'] - ui_quant

print("THANK YOU")
```
